**Load Dataset**

```python
data = pd.read_csv('/content/drive/MyDrive/DSA+ML+DL/Machine Learning/Linear Regression/dataset/Placement_Data.csv')
data.head(5)
```

```python
x = data.iloc[:, :1].values
y = data.iloc[:, 1:].values

# Then create a dataset for training (80% : 20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
class LinearRegressionScratch():
  def __init__(self):
    self.slope = None
    self.intercept = None

  def fit(self, x, y):
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = 0
    denominator = 0
    for i in range(n):
      numerator += (x[i] - x_mean) * (y[i] - y_mean)
      denominator += (x[i] - x_mean) ** 2
    self.slope = numerator / denominator
    self.intercept = y_mean - self.slope * x_mean
    # print("Slope : ", self.slope)
    # print("Intercept : ", self.intercept)

  # Prediction
  def predict(self, x):
    y_pred = []
    for i in x:
      y_pred.append(self.slope * i + self.intercept)
    return y_pred
```

**Train**

```python
lr = LinearRegressionScratch()
lr.fit(x, y)
```

```python
# Slope and Intercept
print(lr.slope)
print(lr.intercept)
```

**Prediction**

```python
# Prediction for whole test data
y_pred = lr.predict(x_test)
print(y_pred)
```
